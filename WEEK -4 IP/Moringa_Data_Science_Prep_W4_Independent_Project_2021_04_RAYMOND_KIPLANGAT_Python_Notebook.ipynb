# <font color='Green'> **Data** **Analysis**

```python
# Identify the most popular hour of the day for picking up a shared electric car 
# (Bluecar) in the city of Paris over the month of April 2018.

x = df[(df['Charging Status']== 'operational')& (df['City']=='Paris')]['Bluecar counter'].value_counts().max()
df.loc[x]['Date_time']
```

<font color='Green'> In the above analysis the most popular hour for renting cars is 8:15 where the cars are maximum at the station and that means the station is supposed to be operational also city should be Paris.

```python
# What is the most popular hour for returning cars?
x=df[(df['Charging Status']== 'operational') & (df['Rental status']== 'operational')]['Bluecar counter'].value_counts().min()
df.loc[x]['Date_time']
```

<font color='Green'>The above analysis shows that the time for returning cars is 07:02   the time when cars for renting are minimun in the station and also point to not is that a station should be operatinal.

```python
# What station is the most popular?
# Overall?
x = df[(df['Charging Status']== 'operational')&(df['Rental status']=='operational')]['Bluecar counter'].value_counts().max()
df.loc[x]['Public name']
```
