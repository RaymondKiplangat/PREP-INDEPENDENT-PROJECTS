```python
# TO JOIN THE TWO TABLES WE NEED TO CREATE AN INNER JOIN.
%%sql
create table Combined  as
SELECT Population.State , Population, GrandElectors
FROM GrandElectors JOIN Population
ON GrandElectors.State = Population.State
ORDER BY Population DESC ;
```

```python
#To order the states by decreasing ratio of Grand Electors per capita. That will make our priority list

%%sql 
SELECT Ratio1
FROM Combined
ORDER BY Ratio1 DESC LIMIT 10;
```

```python
#To compute the running total of Grand Electors in that sorted list

%%sql 
SELECT sum(Ratio1)/Ratio1
FROM Combined

```
