```python
# Let's import the pandas library
import pandas as pd
# as well as the Numpy library
import numpy as np
```

### LOADING TELKOM DATASET 

```python
df1 = pd.read_csv('/content/Telcom_dataset.csv', delimiter = ',')
df1
```

```python
df2 = pd.read_csv('/content/Telcom_dataset2.csv', delimiter = ',')
df2
```

```python
df3 = pd.read_csv('/content/Telcom_dataset3.csv', delimiter = ',')
df3
```

### DATA CLEANING EXAMPLE RENAMING THE PRODUTC  COLUMN TO PRODUCT.
### RENAMING THE DW_A_NUMBER AND DW_B_NUMBER TO DW_A_NUMBER_INT AND DW_B_NUMBER_INT
### RENAMING THE CELLID AND SIET_ID TO CELL_ID AND SITE_ID

```python
df1_N = df1.rename({'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'},axis='columns',inplace=True)
df1_N
```

```python
df2_N = df2.rename({'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'},axis='columns',inplace=True)
df2_N
```

```python
df3_N = df3.rename({'CELLID':'CELL_ID','SIET_ID':'SITE_ID'}, axis ='columns',inplace =True)
df3_N
```

### MERGED
### DataFrame: (df1_N, df2_N, df3_N) looks similar now in way.so we have to merge all the three tables to be one.

```python

df_all = pd.concat((df1,df2,df3), axis = 0)
df_all
```

### Data cleaning removing the (na)  from the row and columns values

```python
df_all = df_all.dropna(axis='index',how ='any')
df_all
```

```python
df_all.drop_duplicates(keep=False)
df_all
```

```python
df_all.drop(['COUNTRY_A','COUNTRY_B','CELL_ON_SITE'],axis = 1,inplace=True)
df_all
```

```python
df_all['PRODUCT'].value_counts()
```

```python
grouped=df_all.groupby(['PRODUCT']).sum()
grouped
```

```python
df_all['PRODUCT']=df_all['PRODUCT'].astype('string')
df_all['DATE_TIME']=df_all['DATE_TIME'].astype('string')
df_all['DW_A_NUMBER_INT']=df_all['DW_A_NUMBER_INT'].astype('string')
df_all['DW_B_NUMBER_INT']=df_all['DW_B_NUMBER_INT'].astype('string')
df_all['CELL_ID']=df_all['CELL_ID'].astype('string')
df_all['SITE_ID']=df_all['SITE_ID'].astype('string')
```
